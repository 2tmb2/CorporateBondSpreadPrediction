```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score,
    ConfusionMatrixDisplay, classification_report,
)
import xgboost as xgb
import sys; sys.path.insert(0, '../..')
from feature_engineering import add_trend_features
```

```python
df = pd.read_csv('../../data/dataset.csv', parse_dates=['Date']).set_index('Date').sort_index()
df = add_trend_features(df)
df.head(3)
```

```python
TRAIN_WINDOW_YEARS = 30
start_eval_year    = 2000
XGB_PARAMS = dict(learning_rate=0.05, max_depth=6, subsample=0.8,
                  colsample_bytree=0.8, random_state=42, n_jobs=-1)
CLASSES = [0, 1, 2]  # DOWN, FLAT, UP -- needed for roc_auc_score when FLAT absent from fold
import warnings as _w

for H in [1, 7, 28]:
    print(f"\n{'='*62}")
    print(f"  H = {H} day(s) ahead")
    print(f"{'='*62}\n")

    # Target construction
    y_delta = df['BAA10Y'].shift(-H) - df['BAA10Y']
    y_dir   = np.sign(y_delta)
    y_now   = df['BAA10Y']
    X_full  = df.drop(columns=['BAA10Y']).copy()
    X_full['BAA10Y_current'] = df['BAA10Y']

    # 3-class: DOWN=0, FLAT=1, UP=2
    mask = y_delta.notna() & y_now.notna() & X_full.notna().all(axis=1)
    X   = X_full.loc[mask]
    y   = (y_dir.loc[mask] + 1).astype(int)
    y_t = y_now.loc[mask]
    print(f"Samples: {len(y)}  DOWN={(y==0).sum()} ({(y==0).mean():.1%})  FLAT={(y==1).sum()} ({(y==1).mean():.1%})  UP={(y==2).sum()} ({(y==2).mean():.1%})")

    # Fold building (rolling 30-year window)
    folds = []
    for year in range(start_eval_year, y.index.year.max() + 1):
        train_start = max(X.index.min(), pd.Timestamp(f"{year - TRAIN_WINDOW_YEARS}-10-01"))
        train_end   = pd.Timestamp(f"{year}-09-30")
        test_start  = pd.Timestamp(f"{year}-10-01")
        test_end    = pd.Timestamp(f"{year}-12-31")
        tr = (X.index >= train_start) & (X.index <= train_end)
        te = (X.index >= test_start)  & (X.index <= test_end)
        if tr.sum() > 0 and te.sum() > 0:
            folds.append({'year': year, 'X_train': X.loc[tr], 'y_train': y.loc[tr],
                          'X_test': X.loc[te], 'y_test': y.loc[te], 'y_t_test': y_t.loc[te]})
    print(f"Folds: {len(folds)}  ({folds[0]['year']} -> {folds[-1]['year']})")

    # Hyperparameter sweep (early stopping on fold-0 train)
    X_sw = folds[0]['X_train']
    y_sw = folds[0]['y_train']
    cut  = int(len(X_sw) * 0.80)
    sweep_model = xgb.XGBClassifier(n_estimators=2000, early_stopping_rounds=50,
                                     eval_metric='mlogloss', **XGB_PARAMS)
    sweep_model.fit(X_sw.iloc[:cut], y_sw.iloc[:cut],
                    eval_set=[(X_sw.iloc[cut:], y_sw.iloc[cut:])], verbose=False)
    # best_iteration is 0-indexed; +1 converts to n_estimators count
    best_n = max(1, sweep_model.best_iteration + 1)
    print(f"Best n_estimators: {best_n}")

    val_ll = sweep_model.evals_result()['validation_0']['mlogloss']
    plt.figure(figsize=(8, 3))
    plt.plot(val_ll)
    plt.axvline(best_n - 1, color='red', linestyle='--', label=f'Best = {best_n}')
    plt.xlabel('Round'); plt.ylabel('Val mlogloss')
    plt.title(f'Validation Log-Loss (H={H})')
    plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

    # Walk-forward evaluation
    fold_results, all_y_test, all_y_pred, all_y_proba = [], [], [], []
    for fold in folds:
        clf = xgb.XGBClassifier(n_estimators=best_n, **XGB_PARAMS)
        clf.fit(fold['X_train'], fold['y_train'])
        y_true  = fold['y_test']
        y_pred  = pd.Series(clf.predict(fold['X_test']), index=y_true.index)
        y_proba = clf.predict_proba(fold['X_test'])
        acc  = accuracy_score(y_true, y_pred)
        prec = precision_score(y_true, y_pred, average='weighted', zero_division=0)
        rec  = recall_score(y_true, y_pred, average='weighted', zero_division=0)
        f1   = f1_score(y_true, y_pred, average='weighted', zero_division=0)
        present = sorted(y_true.unique().tolist())
        if len(present) >= 2:
            with _w.catch_warnings():
                _w.simplefilter('ignore')
                try:
                    auc = roc_auc_score(y_true, y_proba[:, present],
                                        multi_class='ovr', average='macro', labels=present)
                except Exception:
                    auc = float('nan')
        else:
            auc = float('nan')
        naive_acc = y_true.value_counts(normalize=True).max()
        fold_results.append({'year': fold['year'], 'n_test': len(y_true),
                             'accuracy': acc, 'precision': prec, 'recall': rec,
                             'f1': f1, 'auc': auc, 'naive_acc': naive_acc})
        all_y_test.append(y_true); all_y_pred.append(y_pred); all_y_proba.append(y_proba)
        auc_str = f"{auc:.3f}" if not np.isnan(auc) else "nan"
        print(f"  {fold['year']}: Acc={acc:.2%}  Prec={prec:.2%}  Rec={rec:.2%}  F1={f1:.3f}  AUC={auc_str}  (naive={naive_acc:.2%})")

    y_test_all  = pd.concat(all_y_test)
    y_pred_all  = pd.concat(all_y_pred)
    y_proba_all = np.vstack(all_y_proba)
    results_df  = pd.DataFrame(fold_results).set_index('year')
    mean_auc = results_df['auc'].mean()
    print(f"\nMean     Acc={results_df['accuracy'].mean():.2%}  Prec={results_df['precision'].mean():.2%}  Rec={results_df['recall'].mean():.2%}  F1={results_df['f1'].mean():.3f}  AUC={mean_auc:.3f}  (naive={results_df['naive_acc'].mean():.2%})")

    # Classification report
    print(f"\nPooled classification report (H={H}):")
    print(classification_report(y_test_all, y_pred_all, target_names=['DOWN (0)', 'FLAT (1)', 'UP (2)'], zero_division=0))
    try:
        pooled_auc = roc_auc_score(y_test_all, y_proba_all, multi_class='ovr', average='macro', labels=CLASSES)
        print(f"Overall AUC OvR macro: {pooled_auc:.4f}")
    except ValueError as e:
        print(f"Overall AUC: n/a ({e})")

    # Confusion matrix
    fig, ax = plt.subplots(figsize=(6, 5))
    ConfusionMatrixDisplay.from_predictions(y_test_all, y_pred_all, labels=CLASSES,
                                            display_labels=['DOWN', 'FLAT', 'UP'], cmap='Blues', ax=ax)
    ax.set_title(f'Confusion Matrix -- H={H} (pooled)')
    plt.tight_layout(); plt.show()

    # Per-fold accuracy and AUC
    fig, axes = plt.subplots(2, 1, figsize=(13, 7), sharex=True)
    axes[0].bar(results_df.index, results_df['accuracy'], color='steelblue', alpha=0.7)
    axes[0].axhline(0.5, color='red', lw=1, linestyle='--', label='50% baseline')
    axes[0].axhline(results_df['accuracy'].mean(), color='green', lw=1.5, linestyle='--',
                    label=f"Mean {results_df['accuracy'].mean():.2%}")
    axes[0].set_ylabel('Accuracy'); axes[0].legend()
    axes[0].set_title(f'Per-Fold Accuracy & AUC -- H={H} (Oct-Dec test window)')
    axes[0].grid(True, alpha=0.3, axis='y')
    axes[1].bar(results_df.index, results_df['auc'].fillna(0.5), color='darkorange', alpha=0.7)
    axes[1].axhline(0.5, color='red', lw=1, linestyle='--', label='0.5 baseline')
    axes[1].axhline(mean_auc, color='green', lw=1.5, linestyle='--',
                    label=f"Mean {mean_auc:.3f} (NaN->0.5 for display)")
    axes[1].set_ylabel('AUC (OvR macro)'); axes[1].set_xlabel('Year'); axes[1].legend()
    axes[1].grid(True, alpha=0.3, axis='y')
    plt.tight_layout(); plt.show()

    # Feature importance (full-data fit)
    xgb_full = xgb.XGBClassifier(n_estimators=best_n, **XGB_PARAMS)
    xgb_full.fit(X, y)
    importances = pd.Series(xgb_full.feature_importances_, index=X.columns).sort_values()
    fig, ax = plt.subplots(figsize=(8, 10))
    importances.tail(20).plot(kind='barh', ax=ax, color='steelblue', alpha=0.8)
    ax.set_title(f'Top 20 Feature Importances -- H={H} (XGBoost Classifier)')
    ax.set_xlabel('Importance'); ax.grid(True, alpha=0.2, axis='x')
    plt.tight_layout(); plt.show()
```
